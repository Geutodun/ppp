```python
t = np.arange(0, 2*np.pi, 0.01)

plt.figure(figsize=(10,6))
plt.plot(t, np.sin(t), lw=3, label='sin')

#plt.grid()
plt.legend()
plt.title('sin graph')
plt.xlabel('sec')
plt.ylabel('sin value')
plt.show()
```

```python
t = np.arange(0, 2*np.pi, 0.01)

plt.figure(figsize=(10,6))

plt.plot(t, np.cos(t), 'r', label='cos')
plt.grid()
plt.legend()
plt.title('sine')
plt.xlabel('sec')
plt.ylabel('amplitude')
plt.show()
```

```python
t = np.arange(0, 2*np.pi, 0.01)

plt.figure(figsize=(10,6))
plt.plot(t, np.sin(t), lw=3, label='sin')
plt.plot(t, np.cos(t), 'r', label='cos')
plt.grid()
plt.legend()
plt.title('sine')
plt.xlabel('sec')
plt.ylabel('amplitude')
plt.show()
```

```python
t = np.arange(0, 2*np.pi, 0.01)

plt.figure(figsize=(10,6))
plt.plot(t, np.sin(t), lw=10, label='sin')  #lw = 굵기
plt.plot(t, np.cos(t), 'r', label='cos') #'r' 는(=) 빨간색이라는 뜻
plt.grid()
plt.legend()
plt.title('sine')
plt.xlabel('sec')
plt.ylabel('amplitude')
plt.show()
```
